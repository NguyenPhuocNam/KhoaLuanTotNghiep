```python
import os
import json
from pathlib import Path
from collections import defaultdict
import pandas as pd

# ===================================g=========================
# 1. CONFIG PATH
# ============================================================

# Thư mục chứa các file caption JSON
CAPTION_DIR = Path("/kaggle/input/datasets/npnam0123/caption-images")

# Thư mục gốc chứa ảnh thật Fashion-200k
FASHION_ROOT = Path("/kaggle/input/datasets/mayukh18/fashion200k-dataset")

caption_files = [
    "dress_train_captions_vi_google.json",
    "jacket_train_captions_vi_google.json",
    "pants_train_captions_vi_google.json",
    "skirt_train_captions_vi_google.json",
    "top_train_captions_vi_google.json",
]


# ============================================================
# 2. AUTO FIND CAPTION_DIR NẾU PATH TRÊN CHƯA ĐÚNG
# ============================================================

def find_caption_dir(search_root="/kaggle/input"):
    for root, dirs, files in os.walk(search_root):
        if "Caption_images" in dirs:
            return Path(root) / "Caption_images"
    return None


if not CAPTION_DIR.exists():
    found = find_caption_dir("/kaggle/input")
    if found is not None:
        print(f"[INFO] Tự động tìm thấy Caption_images tại: {found}")
        CAPTION_DIR = found
    else:
        print("[WARNING] Không tìm thấy Caption_images trong /kaggle/input")


print("CAPTION_DIR:", CAPTION_DIR)
print("FASHION_ROOT:", FASHION_ROOT)
print("Caption dir exists:", CAPTION_DIR.exists())
print("Fashion root exists:", FASHION_ROOT.exists())


# ============================================================
# 3. HÀM CHUẨN HÓA PATH TRONG FILE CAPTION
# ============================================================

def normalize_caption_path(path_str):
    """
    Chuyển path trong JSON về dạng relative path tính từ FASHION_ROOT.

    Ví dụ:
    f200k_images/pants/straight-leg_pants/90759133/90759133_0.jpeg

    thành:
    women/pants/straight-leg_pants/90759133/90759133_0.jpeg
    """

    p = str(path_str).replace("\\", "/").strip()

    prefixes_to_remove = [
        "Fashion-200k Dataset/",
        "./Fashion-200k Dataset/",
        "/Fashion-200k Dataset/",
        "fashion200k-dataset/",
        "./fashion200k-dataset/",
        "/fashion200k-dataset/",
    ]

    for prefix in prefixes_to_remove:
        if p.startswith(prefix):
            p = p[len(prefix):]

    # Caption cũ thường bắt đầu bằng f200k_images/
    if p.startswith("f200k_images/"):
        p = p[len("f200k_images/"):]

    # Ảnh thật nằm trong women/
    if not p.startswith("women/"):
        p = "women/" + p

    return p


# ============================================================
# 4. INDEX TOÀN BỘ ẢNH THẬT TRONG FASHION-200K
# ============================================================

def build_image_index(root):
    image_exts = {".jpg", ".jpeg", ".png", ".webp"}

    rel_paths = set()
    by_filename = defaultdict(list)
    by_id_filename = defaultdict(list)
    by_category_id_filename = defaultdict(list)

    for img_path in root.rglob("*"):
        if img_path.suffix.lower() in image_exts:
            rel = img_path.relative_to(root).as_posix()
            rel_paths.add(rel)

            filename = img_path.name
            parent_id = img_path.parent.name

            parts = Path(rel).parts

            category = None
            if len(parts) >= 3:
                # women/pants/straight-leg_pants/90759133/90759133_0.jpeg
                category = parts[1]

            by_filename[filename].append(rel)
            by_id_filename[(parent_id, filename)].append(rel)

            if category is not None:
                by_category_id_filename[(category, parent_id, filename)].append(rel)

    return rel_paths, by_filename, by_id_filename, by_category_id_filename


print("\n[INFO] Đang index ảnh thật...")
rel_paths, by_filename, by_id_filename, by_category_id_filename = build_image_index(FASHION_ROOT)
print(f"[INFO] Tổng số ảnh tìm thấy: {len(rel_paths)}")


# ============================================================
# 5. CHECK MỘT FILE CAPTION
# ============================================================

def infer_category_from_path(path_str):
    p = str(path_str).replace("\\", "/")
    parts = p.split("/")

    valid_categories = {"dresses", "jackets", "pants", "skirts", "tops"}

    for part in parts:
        if part in valid_categories:
            return part

    return None


def check_caption_file(json_path):
    with open(json_path, "r", encoding="utf-8") as f:
        captions = json.load(f)

    rows = []

    for raw_path, data in captions.items():
        normalized = normalize_caption_path(raw_path)
        exists = normalized in rel_paths

        suggested_path = None
        match_type = None
        num_candidates = 0

        if exists:
            suggested_path = normalized
            match_type = "exact_after_normalize"
            num_candidates = 1

        else:
            filename = Path(normalized).name
            parent_id = Path(normalized).parent.name
            category = infer_category_from_path(normalized)

            candidates = []

            # Ưu tiên match theo category + id folder + filename
            if category is not None:
                candidates = by_category_id_filename.get((category, parent_id, filename), [])

            if len(candidates) == 1:
                suggested_path = candidates[0]
                match_type = "matched_by_category_id_filename"
                num_candidates = 1

            elif len(candidates) > 1:
                suggested_path = " | ".join(candidates[:10])
                match_type = "multiple_category_id_filename_matches"
                num_candidates = len(candidates)

            else:
                # Match theo id folder + filename
                candidates = by_id_filename.get((parent_id, filename), [])

                if len(candidates) == 1:
                    suggested_path = candidates[0]
                    match_type = "matched_by_id_filename"
                    num_candidates = 1

                elif len(candidates) > 1:
                    suggested_path = " | ".join(candidates[:10])
                    match_type = "multiple_id_filename_matches"
                    num_candidates = len(candidates)

                else:
                    # Match cuối cùng theo filename
                    candidates = by_filename.get(filename, [])

                    if len(candidates) == 1:
                        suggested_path = candidates[0]
                        match_type = "matched_by_filename"
                        num_candidates = 1

                    elif len(candidates) > 1:
                        suggested_path = " | ".join(candidates[:10])
                        match_type = "multiple_filename_matches"
                        num_candidates = len(candidates)

                    else:
                        suggested_path = None
                        match_type = "not_found"
                        num_candidates = 0

        rows.append({
            "caption_file": json_path.name,
            "raw_path": raw_path,
            "normalized_path": normalized,
            "exists": exists,
            "match_type": match_type,
            "num_candidates": num_candidates,
            "suggested_path": suggested_path,
            "absolute_suggested_path": (
                str(FASHION_ROOT / suggested_path)
                if suggested_path is not None and "|" not in suggested_path
                else None
            ),
            "caption_en": data.get("caption_en", ""),
            "caption_vi": data.get("caption_vi", ""),
        })

    return rows


# ============================================================
# 6. CHECK TOÀN BỘ CAPTION FILES
# ============================================================

all_rows = []

for fname in caption_files:
    json_path = CAPTION_DIR / fname

    if not json_path.exists():
        print(f"[WARNING] Không tìm thấy file caption: {json_path}")
        continue

    print(f"[INFO] Đang kiểm tra: {json_path.name}")
    rows = check_caption_file(json_path)
    all_rows.extend(rows)

df = pd.DataFrame(all_rows)

print("\n==============================")
print("KẾT QUẢ TỔNG QUAN")
print("==============================")

print("Tổng số caption:", len(df))

if len(df) > 0:
    print("Số path đúng:", int(df["exists"].sum()))
    print("Số path lỗi:", int((~df["exists"]).sum()))
    print("Tỷ lệ đúng:", round(df["exists"].mean() * 100, 2), "%")

    print("\nThống kê match_type:")
    print(df["match_type"].value_counts())


# ============================================================
# 7. HIỂN THỊ CÁC PATH BỊ LỖI
# ============================================================

bad_df = df[df["exists"] == False].copy()

print("\n==============================")
print("MẪU PATH BỊ LỖI")
print("==============================")

display_cols = [
    "caption_file",
    "raw_path",
    "normalized_path",
    "match_type",
    "num_candidates",
    "suggested_path",
    "absolute_suggested_path",
]

if len(bad_df) > 0:
    display(bad_df[display_cols].head(50))
else:
    print("Không có path lỗi.")


# ============================================================
# 8. LƯU REPORT RA CSV
# ============================================================

output_full = "/kaggle/working/fashion200k_caption_path_check_full.csv"
output_errors = "/kaggle/working/fashion200k_caption_path_errors.csv"

df.to_csv(output_full, index=False, encoding="utf-8-sig")
bad_df.to_csv(output_errors, index=False, encoding="utf-8-sig")

print("\n[INFO] Đã lưu report:")
print(output_full)
print(output_errors)


# ============================================================
# 9. TẠO FILE JSON ĐÃ SỬA PATH
# ============================================================

FIXED_OUTPUT_DIR = Path("/kaggle/working/fixed_caption_json")
FIXED_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)


def create_fixed_caption_json(json_path, rows_df):
    with open(json_path, "r", encoding="utf-8") as f:
        captions = json.load(f)

    fixed_captions = {}
    skipped = {}

    sub_df = rows_df[rows_df["caption_file"] == json_path.name]

    for _, row in sub_df.iterrows():
        raw_path = row["raw_path"]
        suggested_path = row["suggested_path"]
        match_type = row["match_type"]

        # Chỉ sửa khi path match chắc chắn, không có nhiều candidate
        safe_match_types = {
            "exact_after_normalize",
            "matched_by_category_id_filename",
            "matched_by_id_filename",
            "matched_by_filename",
        }

        if match_type in safe_match_types and suggested_path is not None:
            fixed_captions[suggested_path] = captions[raw_path]
        else:
            skipped[raw_path] = {
                "reason": match_type,
                "suggested_path": suggested_path,
                "data": captions[raw_path],
            }

    fixed_path = FIXED_OUTPUT_DIR / json_path.name
    skipped_path = FIXED_OUTPUT_DIR / json_path.name.replace(".json", "_skipped.json")

    with open(fixed_path, "w", encoding="utf-8") as f:
        json.dump(fixed_captions, f, ensure_ascii=False, indent=2)

    with open(skipped_path, "w", encoding="utf-8") as f:
        json.dump(skipped, f, ensure_ascii=False, indent=2)

    return fixed_path, skipped_path, len(fixed_captions), len(skipped)


print("\n==============================")
print("TẠO FILE JSON ĐÃ SỬA PATH")
print("==============================")

for fname in caption_files:
    json_path = CAPTION_DIR / fname

    if not json_path.exists():
        continue

    fixed_path, skipped_path, n_fixed, n_skipped = create_fixed_caption_json(json_path, df)

    print(f"\nFile: {fname}")
    print("Fixed captions:", n_fixed)
    print("Skipped captions:", n_skipped)
    print("Saved fixed:", fixed_path)
    print("Saved skipped:", skipped_path)


# ============================================================
# 10. TEST LOAD ẢNH TỪ PATH ĐÃ SỬA
# ============================================================

from PIL import Image
import matplotlib.pyplot as plt


def show_sample_images(rows_df, n=5):
    sample_df = rows_df[
        rows_df["absolute_suggested_path"].notna()
    ].sample(min(n, len(rows_df)), random_state=42)

    for _, row in sample_df.iterrows():
        img_path = row["absolute_suggested_path"]

        if not os.path.exists(img_path):
            print("[MISSING]", img_path)
            continue

        img = Image.open(img_path)

        plt.figure(figsize=(4, 5))
        plt.imshow(img)
        plt.axis("off")
        plt.title(row["suggested_path"])
        plt.show()

        print("Caption VI:", row["caption_vi"])
        print("-" * 80)


print("\n[INFO] Hiển thị thử vài ảnh đã match:")
show_sample_images(df, n=5)
```

# Thống kê số lượng caption khi tách bằng dấu chấm câu.

```python
import json
import re
from pathlib import Path
import pandas as pd

# ============================================================
# CONFIG
# ============================================================

CAPTION_DIR = Path("/kaggle/working/fixed_caption_json")
# Nếu không dùng file fixed, đổi thành:
# CAPTION_DIR = Path("/kaggle/input/datasets/npnam0123/caption-images")

caption_files = [
    "dress_train_captions_vi_google.json",
    "jacket_train_captions_vi_google.json",
    "pants_train_captions_vi_google.json",
    "skirt_train_captions_vi_google.json",
    "top_train_captions_vi_google.json",
]

# Chọn caption cần thống kê: "caption_en" hoặc "caption_vi"
CAPTION_FIELD = "caption_vi"


# ============================================================
# SPLIT FUNCTION
# ============================================================

def split_by_period(text):
    """
    Tách caption theo dấu chấm.
    Loại bỏ câu rỗng sau khi split.
    """
    if not isinstance(text, str):
        return []

    parts = text.split(".")
    parts = [p.strip() for p in parts if p.strip()]

    return parts


def count_words(text):
    """
    Đếm số từ đơn giản.
    Hỗ trợ cả tiếng Anh và tiếng Việt.
    """
    return len(re.findall(r"\S+", text))


# ============================================================
# COLLECT STATS
# ============================================================

image_rows = []
sentence_rows = []

for fname in caption_files:
    json_path = CAPTION_DIR / fname

    if not json_path.exists():
        print(f"[WARNING] Không tìm thấy: {json_path}")
        continue

    print(f"[INFO] Loading {fname}")

    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    for img_path, captions in data.items():
        caption = captions.get(CAPTION_FIELD, "")

        split_captions = split_by_period(caption)

        image_rows.append({
            "caption_file": fname,
            "image_path": img_path,
            "num_split_captions": len(split_captions),
            "original_caption_words": count_words(caption),
            "original_caption_chars": len(caption),
        })

        for sent_idx, sent in enumerate(split_captions):
            sentence_rows.append({
                "caption_file": fname,
                "image_path": img_path,
                "sentence_index": sent_idx,
                "split_caption": sent,
                "word_length": count_words(sent),
                "char_length": len(sent),
            })


image_df = pd.DataFrame(image_rows)
sentence_df = pd.DataFrame(sentence_rows)


# ============================================================
# SUMMARY
# ============================================================

print("\n==============================")
print(f"THỐNG KÊ CHO FIELD: {CAPTION_FIELD}")
print("==============================")

print("\nTổng số ảnh:", len(image_df))
print("Tổng số caption sau khi tách:", len(sentence_df))

print("\nSố caption con trên mỗi ảnh:")
print("Mean:", image_df["num_split_captions"].mean())
print("Min :", image_df["num_split_captions"].min())
print("Max :", image_df["num_split_captions"].max())

print("\nĐộ dài caption con sau khi tách - theo số từ:")
print("Mean:", sentence_df["word_length"].mean())
print("Min :", sentence_df["word_length"].min())
print("Max :", sentence_df["word_length"].max())

print("\nĐộ dài caption con sau khi tách - theo số ký tự:")
print("Mean:", sentence_df["char_length"].mean())
print("Min :", sentence_df["char_length"].min())
print("Max :", sentence_df["char_length"].max())


# ============================================================
# THỐNG KÊ THEO TỪNG FILE/CATEGORY
# ============================================================

print("\n==============================")
print("THỐNG KÊ THEO FILE")
print("==============================")

file_stats = image_df.groupby("caption_file")["num_split_captions"].agg(
    mean="mean",
    min="min",
    max="max",
    count="count"
)

display(file_stats)

length_stats_by_file = sentence_df.groupby("caption_file")["word_length"].agg(
    mean="mean",
    min="min",
    max="max",
    count="count"
)

display(length_stats_by_file)


# ============================================================
# XEM MỘT VÀI CAPTION CON
# ============================================================

display(sentence_df.head(20))


# ============================================================
# SAVE REPORT
# ============================================================

image_df.to_csv(
    f"/kaggle/working/{CAPTION_FIELD}_split_stats_by_image.csv",
    index=False,
    encoding="utf-8-sig"
)

sentence_df.to_csv(
    f"/kaggle/working/{CAPTION_FIELD}_split_sentences.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\n[INFO] Đã lưu:")
print(f"/kaggle/working/{CAPTION_FIELD}_split_stats_by_image.csv")
print(f"/kaggle/working/{CAPTION_FIELD}_split_sentences.csv")
```

# Thuật toán tách bằng dấu chấm câu.

```python
import json
import re
from pathlib import Path
from collections import Counter

# ============================================================
# CONFIG
# ============================================================

INPUT_JSON_DIR = Path("/kaggle/working/fixed_caption_json")
# Nếu muốn dùng caption gốc:
# INPUT_JSON_DIR = Path("/kaggle/input/datasets/npnam0123/caption-images")

OUTPUT_JSON_DIR = Path("/kaggle/working/split_caption_json_v2")
OUTPUT_JSON_DIR.mkdir(parents=True, exist_ok=True)

caption_files = [
    "dress_train_captions_vi_google.json",
    "jacket_train_captions_vi_google.json",
    "pants_train_captions_vi_google.json",
    "skirt_train_captions_vi_google.json",
    "top_train_captions_vi_google.json",
]

MIN_WORDS_EN = 5
MAX_WORDS_EN = 45

MIN_WORDS_VI = 5
MAX_WORDS_VI = 55

SAVE_MISMATCH = True
SAVE_REJECTED = True


# ============================================================
# TEXT UTILS
# ============================================================

def clean_text(text):
    if not isinstance(text, str):
        return ""

    text = text.strip()
    text = re.sub(r"\s+", " ", text)
    return text


def count_words(text):
    if not isinstance(text, str):
        return 0

    return len(re.findall(r"\S+", text.strip()))


def split_caption_safe(text):
    """
    Split caption an toàn hơn split(".").

    Cơ chế:
    - Split theo dấu chấm có khoảng trắng phía sau: ". "
    - Không split các case như:
      - 3/4 sleeve
      - U.S. size
      - No. 5
      - Mr. / Ms. / Dr.
    """

    text = clean_text(text)

    if not text:
        return []

    # bảo vệ một số abbreviation phổ biến
    protected_patterns = {
        "U.S.": "US_ABBR",
        "U.K.": "UK_ABBR",
        "E.U.": "EU_ABBR",
        "Mr.": "MR_ABBR",
        "Mrs.": "MRS_ABBR",
        "Ms.": "MS_ABBR",
        "Dr.": "DR_ABBR",
        "No.": "NO_ABBR",
    }

    for old, new in protected_patterns.items():
        text = text.replace(old, new)

    # split theo dấu chấm + khoảng trắng
    parts = re.split(r"\.\s+", text)

    sentences = []

    for part in parts:
        part = part.strip()

        # bỏ dấu chấm cuối nếu còn
        part = part.rstrip(".")

        # restore abbreviation
        for old, new in protected_patterns.items():
            part = part.replace(new, old)

        part = clean_text(part)

        if part:
            sentences.append(part)

    return sentences


def filter_sentences(sentences, min_words, max_words):
    accepted = []
    rejected = []

    for sent in sentences:
        wc = count_words(sent)

        if wc < min_words:
            rejected.append({
                "sentence": sent,
                "word_count": wc,
                "reason": "too_short"
            })
        elif wc > max_words:
            rejected.append({
                "sentence": sent,
                "word_count": wc,
                "reason": "too_long"
            })
        else:
            accepted.append(sent)

    return accepted, rejected


# ============================================================
# MAIN
# ============================================================

global_stats = Counter()
file_reports = []

for fname in caption_files:
    input_path = INPUT_JSON_DIR / fname

    if not input_path.exists():
        print(f"[WARNING] Không tìm thấy file: {input_path}")
        global_stats["missing_files"] += 1
        continue

    print(f"[INFO] Đang xử lý: {fname}")

    with open(input_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    output_data = {}
    mismatch_data = {}
    rejected_data = {}

    file_stats = Counter()
    file_stats["file"] = fname

    for image_path, captions in data.items():
        global_stats["total_images"] += 1
        file_stats["total_images"] += 1

        caption_en = clean_text(captions.get("caption_en", ""))
        caption_vi = clean_text(captions.get("caption_vi", ""))

        if not caption_en:
            global_stats["empty_caption_en"] += 1
            file_stats["empty_caption_en"] += 1

        if not caption_vi:
            global_stats["empty_caption_vi"] += 1
            file_stats["empty_caption_vi"] += 1

        raw_split_en = split_caption_safe(caption_en)
        raw_split_vi = split_caption_safe(caption_vi)

        valid_split_en, rejected_en = filter_sentences(
            raw_split_en,
            MIN_WORDS_EN,
            MAX_WORDS_EN
        )

        valid_split_vi, rejected_vi = filter_sentences(
            raw_split_vi,
            MIN_WORDS_VI,
            MAX_WORDS_VI
        )

        file_stats["raw_split_en"] += len(raw_split_en)
        file_stats["raw_split_vi"] += len(raw_split_vi)
        file_stats["valid_split_en"] += len(valid_split_en)
        file_stats["valid_split_vi"] += len(valid_split_vi)
        file_stats["rejected_en"] += len(rejected_en)
        file_stats["rejected_vi"] += len(rejected_vi)

        global_stats["raw_split_en"] += len(raw_split_en)
        global_stats["raw_split_vi"] += len(raw_split_vi)
        global_stats["valid_split_en"] += len(valid_split_en)
        global_stats["valid_split_vi"] += len(valid_split_vi)
        global_stats["rejected_en"] += len(rejected_en)
        global_stats["rejected_vi"] += len(rejected_vi)

        raw_count_match = len(raw_split_en) == len(raw_split_vi)
        valid_count_match = len(valid_split_en) == len(valid_split_vi)

        if not raw_count_match:
            global_stats["raw_mismatch_images"] += 1
            file_stats["raw_mismatch_images"] += 1

        if not valid_count_match:
            global_stats["valid_mismatch_images"] += 1
            file_stats["valid_mismatch_images"] += 1

        paired_split_captions = []

        if valid_count_match:
            for en_sent, vi_sent in zip(valid_split_en, valid_split_vi):
                paired_split_captions.append({
                    "caption_en": en_sent,
                    "caption_vi": vi_sent
                })
        else:
            # Không pair nếu lệch số câu sau lọc
            paired_split_captions = []

        if len(valid_split_en) == 0:
            global_stats["no_valid_split_en_images"] += 1
            file_stats["no_valid_split_en_images"] += 1

        if len(valid_split_vi) == 0:
            global_stats["no_valid_split_vi_images"] += 1
            file_stats["no_valid_split_vi_images"] += 1

        if len(paired_split_captions) == 0:
            global_stats["no_paired_split_images"] += 1
            file_stats["no_paired_split_images"] += 1

        output_data[image_path] = {
            "caption_en": caption_en,
            "caption_vi": caption_vi,

            "split_caption_en": valid_split_en,
            "split_caption_vi": valid_split_vi,

            "num_split_caption_en": len(valid_split_en),
            "num_split_caption_vi": len(valid_split_vi),

            "count_match_raw": raw_count_match,
            "count_match_valid": valid_count_match,

            "paired_split_captions": paired_split_captions,
            "num_paired_split_captions": len(paired_split_captions),
        }

        if SAVE_MISMATCH and (not raw_count_match or not valid_count_match):
            mismatch_data[image_path] = {
                "caption_en": caption_en,
                "caption_vi": caption_vi,

                "raw_split_en": raw_split_en,
                "raw_split_vi": raw_split_vi,

                "valid_split_en": valid_split_en,
                "valid_split_vi": valid_split_vi,

                "num_raw_en": len(raw_split_en),
                "num_raw_vi": len(raw_split_vi),
                "num_valid_en": len(valid_split_en),
                "num_valid_vi": len(valid_split_vi),

                "count_match_raw": raw_count_match,
                "count_match_valid": valid_count_match,
            }

        if SAVE_REJECTED and (rejected_en or rejected_vi):
            rejected_data[image_path] = {
                "rejected_en": rejected_en,
                "rejected_vi": rejected_vi,
            }

    output_path = OUTPUT_JSON_DIR / fname.replace(".json", "_split_paired.json")
    mismatch_path = OUTPUT_JSON_DIR / fname.replace(".json", "_mismatch.json")
    rejected_path = OUTPUT_JSON_DIR / fname.replace(".json", "_rejected.json")

    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(output_data, f, ensure_ascii=False, indent=2)

    if SAVE_MISMATCH:
        with open(mismatch_path, "w", encoding="utf-8") as f:
            json.dump(mismatch_data, f, ensure_ascii=False, indent=2)

    if SAVE_REJECTED:
        with open(rejected_path, "w", encoding="utf-8") as f:
            json.dump(rejected_data, f, ensure_ascii=False, indent=2)

    file_reports.append(file_stats)

    print(f"[DONE] Saved split: {output_path}")
    print(f"[DONE] Saved mismatch: {mismatch_path}")
    print(f"[DONE] Saved rejected: {rejected_path}")


# ============================================================
# FINAL REPORT
# ============================================================

print("\n" + "=" * 90)
print("BÁO CÁO KIỂM TRA SAU KHI SPLIT EN + VI")
print("=" * 90)

print("\n[GLOBAL STATS]")
for k, v in global_stats.items():
    print(f"{k}: {v}")

print("\n[FILE REPORTS]")
for report in file_reports:
    print("-" * 90)
    for k, v in report.items():
        print(f"{k}: {v}")

print("\n" + "=" * 90)
print("ĐÁNH GIÁ LỖI")
print("=" * 90)

has_critical_error = False

if global_stats["missing_files"] > 0:
    has_critical_error = True
    print(f"[ERROR] Có {global_stats['missing_files']} file input bị thiếu.")

if global_stats["total_images"] == 0:
    has_critical_error = True
    print("[ERROR] Không xử lý được ảnh nào.")

if global_stats["valid_split_en"] == 0:
    has_critical_error = True
    print("[ERROR] Không tạo được split_caption_en hợp lệ.")

if global_stats["valid_split_vi"] == 0:
    has_critical_error = True
    print("[ERROR] Không tạo được split_caption_vi hợp lệ.")

if global_stats["raw_mismatch_images"] > 0:
    print(f"[WARNING] Có {global_stats['raw_mismatch_images']} ảnh bị lệch số câu raw EN/VI.")

if global_stats["valid_mismatch_images"] > 0:
    print(f"[WARNING] Có {global_stats['valid_mismatch_images']} ảnh bị lệch số câu sau lọc EN/VI.")

if global_stats["no_paired_split_images"] > 0:
    print(f"[WARNING] Có {global_stats['no_paired_split_images']} ảnh không tạo được paired_split_captions.")

if global_stats["rejected_en"] > 0:
    print(f"[INFO] Có {global_stats['rejected_en']} câu EN bị loại vì quá ngắn hoặc quá dài.")

if global_stats["rejected_vi"] > 0:
    print(f"[INFO] Có {global_stats['rejected_vi']} câu VI bị loại vì quá ngắn hoặc quá dài.")

if not has_critical_error:
    print("[SUCCESS] Hoàn tất split EN/VI. Không có lỗi nghiêm trọng.")

print("\nOutput folder:")
print(OUTPUT_JSON_DIR)
```

# Chia train/test (90/10)

```python
import json
import random
from pathlib import Path
from collections import Counter, defaultdict

# ============================================================
# CONFIG
# ============================================================

INPUT_DIR = Path("/kaggle/working/split_caption_json_v2")

OUTPUT_DIR = Path("/kaggle/working/train_test_split_stratified_by_file")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

split_files = [
    "dress_train_captions_vi_google_split_paired.json",
    "jacket_train_captions_vi_google_split_paired.json",
    "pants_train_captions_vi_google_split_paired.json",
    "skirt_train_captions_vi_google_split_paired.json",
    "top_train_captions_vi_google_split_paired.json",
]

TRAIN_RATIO = 0.9
TEST_RATIO = 0.1
RANDOM_SEED = 42

random.seed(RANDOM_SEED)


# ============================================================
# HELPERS
# ============================================================

def count_pairs(data_dict):
    total = 0
    for _, item in data_dict.items():
        total += len(item.get("paired_split_captions", []))
    return total


def get_category_from_filename(fname):
    if fname.startswith("dress"):
        return "dress"
    if fname.startswith("jacket"):
        return "jacket"
    if fname.startswith("pants"):
        return "pants"
    if fname.startswith("skirt"):
        return "skirt"
    if fname.startswith("top"):
        return "top"
    return "unknown"


# ============================================================
# SPLIT RIÊNG TỪNG FILE
# ============================================================

train_data = {}
test_data = {}

train_ids = set()
test_ids = set()

file_reports = []

global_stats = Counter()

for fname in split_files:
    json_path = INPUT_DIR / fname
    category = get_category_from_filename(fname)

    if not json_path.exists():
        print(f"[WARNING] Không tìm thấy: {json_path}")
        global_stats["missing_files"] += 1
        continue

    print(f"[INFO] Loading: {fname}")

    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    # Chỉ lấy image có paired captions hợp lệ
    valid_items = {
        image_path: item
        for image_path, item in data.items()
        if len(item.get("paired_split_captions", [])) > 0
    }

    invalid_count = len(data) - len(valid_items)

    image_paths = list(valid_items.keys())
    random.shuffle(image_paths)

    num_total = len(image_paths)
    num_train = int(num_total * TRAIN_RATIO)

    file_train_ids = set(image_paths[:num_train])
    file_test_ids = set(image_paths[num_train:])

    file_train_data = {}
    file_test_data = {}

    for image_path, item in valid_items.items():
        item["category"] = category
        item["source_file"] = fname

        if image_path in file_train_ids:
            file_train_data[image_path] = item
            train_data[image_path] = item
            train_ids.add(image_path)

        elif image_path in file_test_ids:
            file_test_data[image_path] = item
            test_data[image_path] = item
            test_ids.add(image_path)

    report = {
        "file": fname,
        "category": category,
        "original_images": len(data),
        "valid_images": num_total,
        "invalid_no_paired_images": invalid_count,
        "train_images": len(file_train_data),
        "test_images": len(file_test_data),
        "train_ratio": len(file_train_data) / num_total if num_total > 0 else 0,
        "test_ratio": len(file_test_data) / num_total if num_total > 0 else 0,
        "train_pairs": count_pairs(file_train_data),
        "test_pairs": count_pairs(file_test_data),
    }

    file_reports.append(report)

    global_stats["original_images"] += len(data)
    global_stats["valid_images"] += num_total
    global_stats["invalid_no_paired_images"] += invalid_count
    global_stats["train_images"] += len(file_train_data)
    global_stats["test_images"] += len(file_test_data)
    global_stats["train_pairs"] += report["train_pairs"]
    global_stats["test_pairs"] += report["test_pairs"]

    # Lưu riêng từng category nếu muốn debug
    category_train_path = OUTPUT_DIR / f"{category}_train.json"
    category_test_path = OUTPUT_DIR / f"{category}_test.json"

    with open(category_train_path, "w", encoding="utf-8") as f:
        json.dump(file_train_data, f, ensure_ascii=False, indent=2)

    with open(category_test_path, "w", encoding="utf-8") as f:
        json.dump(file_test_data, f, ensure_ascii=False, indent=2)

    print(f"[DONE] {category}: train={len(file_train_data)}, test={len(file_test_data)}")


# ============================================================
# SAVE MERGED TRAIN / TEST
# ============================================================

train_json_path = OUTPUT_DIR / "fashion_clip_train.json"
test_json_path = OUTPUT_DIR / "fashion_clip_test.json"

with open(train_json_path, "w", encoding="utf-8") as f:
    json.dump(train_data, f, ensure_ascii=False, indent=2)

with open(test_json_path, "w", encoding="utf-8") as f:
    json.dump(test_data, f, ensure_ascii=False, indent=2)


# ============================================================
# SAVE REPORT
# ============================================================

report_path = OUTPUT_DIR / "split_report.json"

report_json = {
    "config": {
        "train_ratio": TRAIN_RATIO,
        "test_ratio": TEST_RATIO,
        "random_seed": RANDOM_SEED,
        "strategy": "split_each_file_then_merge",
    },
    "global_stats": dict(global_stats),
    "file_reports": file_reports,
}

with open(report_path, "w", encoding="utf-8") as f:
    json.dump(report_json, f, ensure_ascii=False, indent=2)


# ============================================================
# VERIFY
# ============================================================

print("\n" + "=" * 90)
print("VERIFY TRAIN / TEST SPLIT")
print("=" * 90)

intersection = train_ids.intersection(test_ids)

if len(intersection) == 0:
    print("[SUCCESS] Không có image leakage giữa train/test.")
else:
    print(f"[ERROR] Có {len(intersection)} image bị overlap giữa train/test.")

print("\n[GLOBAL STATS]")
for k, v in global_stats.items():
    print(f"{k}: {v}")

total_valid = global_stats["valid_images"]

if total_valid > 0:
    print("\nOverall ratio:")
    print("Train:", round(global_stats["train_images"] / total_valid, 4))
    print("Test :", round(global_stats["test_images"] / total_valid, 4))

print("\n[FILE REPORTS]")
for r in file_reports:
    print("-" * 90)
    print("File:", r["file"])
    print("Category:", r["category"])
    print("Original images:", r["original_images"])
    print("Valid images:", r["valid_images"])
    print("Invalid no paired images:", r["invalid_no_paired_images"])
    print("Train images:", r["train_images"])
    print("Test images:", r["test_images"])
    print("Train ratio:", round(r["train_ratio"], 4))
    print("Test ratio:", round(r["test_ratio"], 4))
    print("Train sentence pairs:", r["train_pairs"])
    print("Test sentence pairs:", r["test_pairs"])


# ============================================================
# CATEGORY DISTRIBUTION CHECK
# ============================================================

def category_distribution(data_dict):
    counter = Counter()

    for _, item in data_dict.items():
        counter[item.get("category", "unknown")] += 1

    return counter


train_dist = category_distribution(train_data)
test_dist = category_distribution(test_data)

print("\n" + "=" * 90)
print("CATEGORY DISTRIBUTION")
print("=" * 90)

print("\nTrain distribution:")
for cat, count in train_dist.items():
    print(cat, count)

print("\nTest distribution:")
for cat, count in test_dist.items():
    print(cat, count)


# ============================================================
# SHOW SAMPLE
# ============================================================

if len(train_data) > 0:
    sample_key = random.choice(list(train_data.keys()))
    sample_item = train_data[sample_key]

    print("\n" + "=" * 90)
    print("TRAIN SAMPLE")
    print("=" * 90)

    print("\nImage path:")
    print(sample_key)

    print("\nCategory:")
    print(sample_item.get("category"))

    print("\nSource file:")
    print(sample_item.get("source_file"))

    print("\nNumber of paired captions:")
    print(sample_item.get("num_paired_split_captions"))

    print("\nExample paired captions:\n")

    for idx, pair in enumerate(sample_item.get("paired_split_captions", [])[:3], 1):
        print(f"--- Pair {idx} ---")
        print("[EN]")
        print(pair["caption_en"])
        print("\n[VI]")
        print(pair["caption_vi"])
        print("-" * 60)


print("\n[DONE] Saved files:")
print(train_json_path)
print(test_json_path)
print(report_path)
print("\nOutput folder:")
print(OUTPUT_DIR)
```

# **pre-compute FashionCLIP embeddings**

```python
# ============================================================
# INSTALL
# ============================================================

!pip install -q transformers accelerate tqdm pillow


# ============================================================
# IMPORTS
# ============================================================

import os
import json
import gc
from pathlib import Path

import torch
import torch.nn.functional as F
from PIL import Image, ImageFile
from tqdm.auto import tqdm
from transformers import CLIPProcessor, CLIPModel

ImageFile.LOAD_TRUNCATED_IMAGES = True
os.environ["TOKENIZERS_PARALLELISM"] = "false"


# ============================================================
# CONFIG
# ============================================================

TRAIN_JSON = Path("/kaggle/working/train_test_split_stratified_by_file/fashion_clip_train.json")
FASHION_ROOT = Path("/kaggle/input/datasets/mayukh18/fashion200k-dataset")

OUTPUT_DIR = Path("/kaggle/working/precomputed_fashionclip_t4x2")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

CHUNK_DIR = OUTPUT_DIR / "chunks"
CHUNK_DIR.mkdir(parents=True, exist_ok=True)

MODEL_NAME = "patrickjohncyh/fashion-clip"

IMAGE_BATCH_SIZE = 128
TEXT_BATCH_SIZE = 1024

SAVE_DTYPE = torch.float16
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

print("PyTorch:", torch.__version__)
print("Device:", DEVICE)
print("CUDA available:", torch.cuda.is_available())
print("GPU count:", torch.cuda.device_count())

if torch.cuda.is_available():
    for i in range(torch.cuda.device_count()):
        print(f"GPU {i}:", torch.cuda.get_device_name(i))
        print(f"Capability {i}:", torch.cuda.get_device_capability(i))


# ============================================================
# LOAD TRAIN DATA
# ============================================================

with open(TRAIN_JSON, "r", encoding="utf-8") as f:
    train_data = json.load(f)

image_paths = []
text_records = []
text_list = []

for image_path, item in train_data.items():
    paired = item.get("paired_split_captions", [])

    if len(paired) == 0:
        continue

    image_idx = len(image_paths)
    image_paths.append(image_path)

    for pair_idx, pair in enumerate(paired):
        caption_en = pair.get("caption_en", "").strip()

        if caption_en:
            text_records.append({
                "image_idx": image_idx,
                "image_path": image_path,
                "pair_idx": pair_idx,
                "caption_en": caption_en,
            })
            text_list.append(caption_en)

print("\n==============================")
print("DATA STATS")
print("==============================")
print("Images:", len(image_paths))
print("English captions:", len(text_list))


# ============================================================
# LOAD FASHIONCLIP
# ============================================================

processor = CLIPProcessor.from_pretrained(MODEL_NAME)
model = CLIPModel.from_pretrained(MODEL_NAME)

model = model.to(DEVICE)
model.eval()

print("\nFashionCLIP loaded.")


# ============================================================
# HELPER FUNCTIONS
# ============================================================

def load_image_safe(path):
    try:
        return Image.open(path).convert("RGB")
    except Exception:
        return None


def extract_tensor_output(output):
    if torch.is_tensor(output):
        return output

    if hasattr(output, "image_embeds") and output.image_embeds is not None:
        return output.image_embeds

    if hasattr(output, "text_embeds") and output.text_embeds is not None:
        return output.text_embeds

    if hasattr(output, "pooler_output") and output.pooler_output is not None:
        return output.pooler_output

    if hasattr(output, "last_hidden_state"):
        return output.last_hidden_state[:, 0, :]

    if isinstance(output, (tuple, list)):
        for x in output:
            if torch.is_tensor(x):
                return x

    raise TypeError(f"Không thể trích tensor từ output type: {type(output)}")


def encode_images(pixel_values):
    output = model.get_image_features(pixel_values=pixel_values)
    feats = extract_tensor_output(output)

    # Nếu lỡ ra 3D thì lấy CLS
    if feats.dim() == 3:
        feats = feats[:, 0, :]

    feats = F.normalize(feats, dim=-1)
    return feats


def encode_text(input_ids, attention_mask):
    output = model.get_text_features(
        input_ids=input_ids,
        attention_mask=attention_mask
    )
    feats = extract_tensor_output(output)

    if feats.dim() == 3:
        feats = feats[:, 0, :]

    feats = F.normalize(feats, dim=-1)
    return feats


# ============================================================
# QUICK FORWARD TEST
# ============================================================

print("\n==============================")
print("FASHIONCLIP FORWARD TEST")
print("==============================")

test_img_path = FASHION_ROOT / image_paths[0]
test_img = Image.open(test_img_path).convert("RGB")
test_text = text_list[0]

test_img_inputs = processor(images=[test_img], return_tensors="pt")
test_pixel_values = test_img_inputs["pixel_values"].to(DEVICE)

test_txt_inputs = processor(
    text=[test_text],
    return_tensors="pt",
    padding=True,
    truncation=True,
    max_length=77
)

test_input_ids = test_txt_inputs["input_ids"].to(DEVICE)
test_attention_mask = test_txt_inputs["attention_mask"].to(DEVICE)

with torch.no_grad():
    test_img_feat = encode_images(test_pixel_values)
    test_txt_feat = encode_text(test_input_ids, test_attention_mask)

print("Test image feature:", test_img_feat.shape)
print("Test text feature :", test_txt_feat.shape)
print("[SUCCESS] Forward test OK.")

del test_img, test_pixel_values, test_input_ids, test_attention_mask
del test_img_feat, test_txt_feat
torch.cuda.empty_cache()
gc.collect()


# ============================================================
# IMAGE EMBEDDINGS
# ============================================================

print("\n==============================")
print("EXTRACT IMAGE EMBEDDINGS")
print("==============================")

image_chunk_files = []
valid_image_paths = []
bad_images = []

for chunk_id, start in enumerate(tqdm(range(0, len(image_paths), IMAGE_BATCH_SIZE))):
    batch_paths = image_paths[start:start + IMAGE_BATCH_SIZE]

    images = []
    batch_valid_paths = []

    for rel_path in batch_paths:
        abs_path = FASHION_ROOT / rel_path
        img = load_image_safe(abs_path)

        if img is None:
            bad_images.append(rel_path)
            continue

        images.append(img)
        batch_valid_paths.append(rel_path)

    if len(images) == 0:
        continue

    inputs = processor(images=images, return_tensors="pt")
    pixel_values = inputs["pixel_values"].to(DEVICE)

    with torch.no_grad():
        feats = encode_images(pixel_values)
        feats = feats.detach().cpu().to(SAVE_DTYPE)

    chunk_path = CHUNK_DIR / f"image_embeds_chunk_{chunk_id:05d}.pt"

    torch.save(
        {
            "start": start,
            "batch_paths": batch_valid_paths,
            "embeddings": feats,
        },
        chunk_path
    )

    image_chunk_files.append(str(chunk_path))
    valid_image_paths.extend(batch_valid_paths)

    del inputs, pixel_values, feats, images
    torch.cuda.empty_cache()
    gc.collect()

print("Image chunks:", len(image_chunk_files))
print("Valid images:", len(valid_image_paths))
print("Bad images:", len(bad_images))


# ============================================================
# FILTER TEXT RECORDS IF BAD IMAGES EXIST
# ============================================================

if len(bad_images) > 0:
    print("\n[WARNING] Bad images found. Filtering text records...")

    valid_image_set = set(valid_image_paths)
    old_to_new_idx = {
        path: idx
        for idx, path in enumerate(valid_image_paths)
    }

    new_text_records = []
    new_text_list = []

    for rec in text_records:
        img_path = rec["image_path"]

        if img_path in valid_image_set:
            new_rec = dict(rec)
            new_rec["image_idx"] = old_to_new_idx[img_path]
            new_text_records.append(new_rec)
            new_text_list.append(rec["caption_en"])

    text_records = new_text_records
    text_list = new_text_list

    print("Filtered English captions:", len(text_list))


# ============================================================
# TEXT EMBEDDINGS
# ============================================================

print("\n==============================")
print("EXTRACT TEXT EMBEDDINGS")
print("==============================")

text_chunk_files = []

for chunk_id, start in enumerate(tqdm(range(0, len(text_list), TEXT_BATCH_SIZE))):
    batch_texts = text_list[start:start + TEXT_BATCH_SIZE]

    inputs = processor(
        text=batch_texts,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=77
    )

    input_ids = inputs["input_ids"].to(DEVICE)
    attention_mask = inputs["attention_mask"].to(DEVICE)

    with torch.no_grad():
        feats = encode_text(input_ids, attention_mask)
        feats = feats.detach().cpu().to(SAVE_DTYPE)

    chunk_path = CHUNK_DIR / f"text_embeds_chunk_{chunk_id:05d}.pt"

    torch.save(
        {
            "start": start,
            "texts": batch_texts,
            "embeddings": feats,
        },
        chunk_path
    )

    text_chunk_files.append(str(chunk_path))

    del inputs, input_ids, attention_mask, feats
    torch.cuda.empty_cache()
    gc.collect()

print("Text chunks:", len(text_chunk_files))


# ============================================================
# MERGE CHUNKS
# ============================================================

print("\n==============================")
print("MERGE IMAGE CHUNKS")
print("==============================")

image_embeds_list = []

for chunk_file in tqdm(image_chunk_files):
    obj = torch.load(chunk_file, map_location="cpu")
    image_embeds_list.append(obj["embeddings"])

image_embeddings = torch.cat(image_embeds_list, dim=0)

del image_embeds_list
gc.collect()

print("Image embeddings:", image_embeddings.shape)


print("\n==============================")
print("MERGE TEXT CHUNKS")
print("==============================")

text_embeds_list = []

for chunk_file in tqdm(text_chunk_files):
    obj = torch.load(chunk_file, map_location="cpu")
    text_embeds_list.append(obj["embeddings"])

text_embeddings = torch.cat(text_embeds_list, dim=0)

del text_embeds_list
gc.collect()

print("Text embeddings:", text_embeddings.shape)


# ============================================================
# SAVE FINAL .PT
# ============================================================

output_pt = OUTPUT_DIR / "fashionclip_train_precomputed.pt"

save_obj = {
    "model_name": MODEL_NAME,

    "image_embeddings": image_embeddings,
    "text_embeddings": text_embeddings,

    "image_paths": valid_image_paths,
    "text_records": text_records,

    "bad_images": bad_images,

    "embedding_dim": image_embeddings.shape[-1],
    "image_count": len(valid_image_paths),
    "text_count": len(text_records),

    "save_dtype": str(SAVE_DTYPE),
    "image_batch_size": IMAGE_BATCH_SIZE,
    "text_batch_size": TEXT_BATCH_SIZE,

    "image_chunk_files": image_chunk_files,
    "text_chunk_files": text_chunk_files,
}

torch.save(save_obj, output_pt)

print("\n==============================")
print("SAVED FINAL FILE")
print("==============================")
print(output_pt)


# ============================================================
# VERIFY FINAL FILE
# ============================================================

loaded = torch.load(output_pt, map_location="cpu")

print("\n==============================")
print("VERIFY SAVED FILE")
print("==============================")
print("Image embeddings:", loaded["image_embeddings"].shape)
print("Text embeddings :", loaded["text_embeddings"].shape)
print("Image paths     :", len(loaded["image_paths"]))
print("Text records    :", len(loaded["text_records"]))
print("Bad images      :", len(loaded["bad_images"]))

ok = True

if loaded["image_embeddings"].shape[0] != len(loaded["image_paths"]):
    ok = False
    print("[ERROR] image_embeddings không khớp image_paths.")

if loaded["text_embeddings"].shape[0] != len(loaded["text_records"]):
    ok = False
    print("[ERROR] text_embeddings không khớp text_records.")

if loaded["image_embeddings"].shape[1] != loaded["text_embeddings"].shape[1]:
    ok = False
    print("[ERROR] image/text embedding dim không khớp.")

if ok:
    print("[SUCCESS] Pre-compute FashionCLIP embeddings hoàn tất và hợp lệ.")
```

# FineTune-FashionCLIP-PhoBERT(baseline)

```python
# ============================================================
# FINE-TUNE FASHIONCLIP TEXT ENCODER (STABLE VERSION)
# ============================================================

import os
import json
import random
import gc
from pathlib import Path
import matplotlib.pyplot as plt # Đã thêm thư viện vẽ biểu đồ

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from tqdm.auto import tqdm
from transformers import CLIPProcessor, CLIPModel, get_cosine_schedule_with_warmup

# --- CONFIG ---
TRAIN_JSON = Path("/kaggle/working/train_test_split_stratified_by_file/fashion_clip_train.json")
PRECOMPUTED_PT = Path("/kaggle/working/precomputed_fashionclip_t4x2/fashionclip_train_precomputed.pt")
OUTPUT_DIR = Path("/kaggle/working/fashionclip_direct_distill")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

MODEL_NAME = "patrickjohncyh/fashion-clip"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

MAX_LENGTH = 77   
BATCH_SIZE = 128
NUM_EPOCHS = 3
LR = 2e-5
WEIGHT_DECAY = 0.01
WARMUP_RATIO = 0.05
GRAD_CLIP = 1.0

def set_seed(seed):
    random.seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
set_seed(42)

# --- 1. LOAD DATA (OPTIMIZED) ---
print("[INFO] Loading precomputed embeddings...")
precomputed = torch.load(PRECOMPUTED_PT, map_location="cpu")
teacher_text_embeddings = precomputed["text_embeddings"].float()
text_records = precomputed["text_records"]
del precomputed
gc.collect()

print("[INFO] Building vi_map...")
with open(TRAIN_JSON, "r", encoding="utf-8") as f:
    train_json = json.load(f)

vi_map = {}
for img_p, item in train_json.items():
    for i, pair in enumerate(item.get("paired_split_captions", [])):
        vi_map[(img_p, i)] = pair.get("caption_vi", "").strip()
del train_json
gc.collect()

train_records = []
for i, r in enumerate(text_records):
    key = (r["image_path"], r["pair_idx"])
    if key in vi_map:
        train_records.append({"text": vi_map[key], "t_idx": i})
del vi_map
gc.collect()

# --- 2. DATASET & LOSS ---
class DirectDistillDataset(Dataset):
    def __init__(self, records, teacher_embeds, processor, max_len):
        self.records, self.teacher_embeds, self.processor, self.max_len = records, teacher_embeds, processor, max_len
    def __len__(self): return len(self.records)
    def __getitem__(self, idx):
        rec = self.records[idx]
        enc = self.processor(text=[rec["text"]], padding="max_length", truncation=True, max_length=self.max_len, return_tensors="pt")
        return {"ids": enc["input_ids"].squeeze(0), "mask": enc["attention_mask"].squeeze(0), "target": self.teacher_embeds[rec["t_idx"]]}

def distill_loss(s_emb, t_emb):
    # s_emb lúc này chắc chắn đã là Tensor 2D [Batch, 512]
    t_emb = F.normalize(t_emb, p=2, dim=-1)
    s_emb = F.normalize(s_emb, p=2, dim=-1)
    
    mse = F.mse_loss(s_emb, t_emb)
    cos = 1 - F.cosine_similarity(s_emb, t_emb, dim=-1).mean()
    return mse + cos, mse.detach(), cos.detach()

# --- 3. INIT MODEL ---
print("[INFO] Loading FashionCLIP...")
processor = CLIPProcessor.from_pretrained(MODEL_NAME)
model = CLIPModel.from_pretrained(MODEL_NAME).to(DEVICE)

# Fix vision model
for param in model.vision_model.parameters(): param.requires_grad = False

if torch.cuda.device_count() > 1:
    model = nn.DataParallel(model)

optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
train_loader = DataLoader(DirectDistillDataset(train_records, teacher_text_embeddings, processor, MAX_LENGTH), 
                          batch_size=BATCH_SIZE, shuffle=True, num_workers=0)

total_steps = len(train_loader) * NUM_EPOCHS
scheduler = get_cosine_schedule_with_warmup(optimizer, int(total_steps * WARMUP_RATIO), total_steps)
scaler = torch.amp.GradScaler('cuda')

# --- 4. TRAINING LOOP ---
print(f"[START] Training for {NUM_EPOCHS} epochs...")
train_losses = [] # Đã thêm list lưu loss để vẽ biểu đồ

for epoch in range(NUM_EPOCHS):
    model.train()
    running_loss, running_mse, running_cos = 0.0, 0.0, 0.0
    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS}")

    for step, batch in enumerate(pbar):
        ids, mask, targets = batch["ids"].to(DEVICE), batch["mask"].to(DEVICE), batch["target"].to(DEVICE)
        
        optimizer.zero_grad(set_to_none=True)
        with torch.amp.autocast('cuda'):
            # Forward pass
            outputs = model.module.get_text_features(ids, mask) if hasattr(model, 'module') else model.get_text_features(ids, mask)
            
            # TRÍCH XUẤT TENSOR AN TOÀN (Sửa lỗi AttributeError tại đây)
            if hasattr(outputs, "text_embeds"):
                s_emb = outputs.text_embeds
            elif torch.is_tensor(outputs):
                s_emb = outputs
            else:
                s_emb = outputs[0]
                
            # Đảm bảo s_emb là 2D (Nếu là 3D [Batch, Seq, Dim] thì lấy CLS/EOT)
            if s_emb.dim() == 3:
                s_emb = s_emb[:, 0, :]

            # Tính loss
            loss, mse_v, cos_v = distill_loss(s_emb, targets)

        scaler.scale(loss).backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), GRAD_CLIP)
        scaler.step(optimizer)
        scaler.update()
        scheduler.step()

        running_loss += loss.item()
        running_mse += mse_v.item()
        running_cos += cos_v.item()
        pbar.set_postfix({"loss": f"{running_loss/(step+1):.4f}", "mse": f"{running_mse/(step+1):.4f}"})

    # Đã thêm phần tính và lưu trung bình loss mỗi epoch
    epoch_loss = running_loss / len(train_loader)
    train_losses.append(epoch_loss)

    # Save Checkpoint
    checkpoint_path = OUTPUT_DIR / f"clip_direct_epoch_{epoch+1}.pt"
    torch.save({
        "model_state_dict": model.module.state_dict() if hasattr(model, 'module') else model.state_dict()
    }, checkpoint_path)

print(f"[SUCCESS] Training complete. Models saved in {OUTPUT_DIR}")

# ============================================================
# PLOT LOSS
# ============================================================
print("[INFO] Plotting training loss...")

plt.figure(figsize=(10, 6))
plt.plot(range(1, NUM_EPOCHS + 1), train_losses, marker='o', linestyle='-', color='b', linewidth=2, markersize=8)
plt.title('Training Loss over Epochs', fontsize=14)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.xticks(range(1, NUM_EPOCHS + 1))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

plot_path = OUTPUT_DIR / "training_loss_plot.png"
plt.savefig(plot_path)
plt.show()

print(f"[DONE] Saved loss plot to: {plot_path}")
```

```python
# ============================================================
# INSTALL
# ============================================================

!pip install -q transformers accelerate tqdm py_vncorenlp matplotlib


# ============================================================
# IMPORTS
# ============================================================

import os
import json
import math
import random
from pathlib import Path
import matplotlib.pyplot as plt
import py_vncorenlp

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from tqdm.auto import tqdm
from transformers import AutoTokenizer, AutoModel, get_cosine_schedule_with_warmup


# ============================================================
# CONFIG
# ============================================================

TRAIN_JSON = Path("/kaggle/working/train_test_split_stratified_by_file/fashion_clip_train.json")
TEST_JSON  = Path("/kaggle/working/train_test_split_stratified_by_file/fashion_clip_test.json")

PRECOMPUTED_PT = Path("/kaggle/working/precomputed_fashionclip_t4x2/fashionclip_train_precomputed.pt")

OUTPUT_DIR = Path("/kaggle/working/phobert_fashionclip_student")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

PHOBERT_NAME = "vinai/phobert-base-v2"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

MAX_LENGTH = 96
BATCH_SIZE = 128
NUM_EPOCHS = 3
LR = 2e-5
WEIGHT_DECAY = 0.01
WARMUP_RATIO = 0.05
GRAD_CLIP = 1.0

FREEZE_PHOBERT = False

SEED = 42

print("Device:", DEVICE)
print("GPU count:", torch.cuda.device_count())


# ============================================================
# SEED
# ============================================================

def set_seed(seed):
    random.seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(SEED)


# ============================================================
# LOAD PRECOMPUTED TEACHER EMBEDDINGS
# ============================================================

precomputed = torch.load(PRECOMPUTED_PT, map_location="cpu")

teacher_text_embeddings = precomputed["text_embeddings"].float()
text_records = precomputed["text_records"]

print("Teacher text embeddings:", teacher_text_embeddings.shape)
print("Text records:", len(text_records))


# ============================================================
# SETUP VNCORENLP
# ============================================================
print("Downloading and loading VnCoreNLP...")
py_vncorenlp.download_model(save_dir='/kaggle/working')
rdrsegmenter = py_vncorenlp.VnCoreNLP(annotators=["wseg"], save_dir='/kaggle/working')


# ============================================================
# LOAD JSON TO GET VI CAPTIONS
# ============================================================

with open(TRAIN_JSON, "r", encoding="utf-8") as f:
    train_json = json.load(f)

# Map: image_path + pair_idx -> caption_vi
vi_map = {}

for image_path, item in train_json.items():
    pairs = item.get("paired_split_captions", [])

    for pair_idx, pair in enumerate(pairs):
        key = (image_path, pair_idx)
        vi_map[key] = pair.get("caption_vi", "").strip()


# ============================================================
# BUILD TRAIN RECORDS
# ============================================================

train_records = []
missing_vi = 0

for embed_idx, rec in enumerate(text_records):
    image_path = rec["image_path"]
    pair_idx = rec["pair_idx"]

    vi_caption = vi_map.get((image_path, pair_idx), "")

    if not vi_caption:
        missing_vi += 1
        continue
        
    # Tách từ bằng VnCoreNLP
    segmented_sentences = rdrsegmenter.word_segment(vi_caption)
    # word_segment trả về list các câu, gộp lại bằng khoảng trắng
    vi_caption = " ".join(segmented_sentences)

    train_records.append({
        "caption_vi": vi_caption,
        "teacher_idx": embed_idx,
        "image_path": image_path,
        "pair_idx": pair_idx,
    })

print("Train records:", len(train_records))
print("Missing VI:", missing_vi)


# ============================================================
# DATASET
# ============================================================

class DistillDataset(Dataset):
    def __init__(self, records, teacher_embeddings, tokenizer, max_length):
        self.records = records
        self.teacher_embeddings = teacher_embeddings
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.records)

    def __getitem__(self, idx):
        item = self.records[idx]

        text = item["caption_vi"]
        teacher_idx = item["teacher_idx"]

        enc = self.tokenizer(
            text,
            padding="max_length",
            truncation=True,
            max_length=self.max_length,
            return_tensors="pt"
        )

        return {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
            "teacher_embedding": self.teacher_embeddings[teacher_idx],
        }


# ============================================================
# MODEL
# ============================================================

class PhoBERTFashionCLIPStudent(nn.Module):
    def __init__(self, phobert_name="vinai/phobert-base-v2", out_dim=512, freeze_phobert=False):
        super().__init__()

        self.encoder = AutoModel.from_pretrained(phobert_name)
        hidden_size = self.encoder.config.hidden_size

        self.projection = nn.Sequential(
            nn.Linear(hidden_size, hidden_size),
            nn.GELU(),
            nn.LayerNorm(hidden_size),
            nn.Dropout(0.1),
            nn.Linear(hidden_size, out_dim),
        )

        if freeze_phobert:
            for p in self.encoder.parameters():
                p.requires_grad = False

    def forward(self, input_ids, attention_mask):
        outputs = self.encoder(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        cls = outputs.last_hidden_state[:, 0, :]
        emb = self.projection(cls)
        emb = F.normalize(emb, dim=-1)

        return emb


# ============================================================
# TOKENIZER / DATALOADER
# ============================================================

tokenizer = AutoTokenizer.from_pretrained(PHOBERT_NAME)

train_dataset = DistillDataset(
    records=train_records,
    teacher_embeddings=teacher_text_embeddings,
    tokenizer=tokenizer,
    max_length=MAX_LENGTH
)

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2,
    pin_memory=True,
    drop_last=True
)

print("Train batches:", len(train_loader))


# ============================================================
# INIT MODEL
# ============================================================

student = PhoBERTFashionCLIPStudent(
    phobert_name=PHOBERT_NAME,
    out_dim=teacher_text_embeddings.shape[1],
    freeze_phobert=FREEZE_PHOBERT
)

student = student.to(DEVICE)

if torch.cuda.device_count() > 1:
    student = nn.DataParallel(student)
    print("Using DataParallel for student.")


# ============================================================
# OPTIMIZER / SCHEDULER
# ============================================================

optimizer = torch.optim.AdamW(
    student.parameters(),
    lr=LR,
    weight_decay=WEIGHT_DECAY
)

total_steps = len(train_loader) * NUM_EPOCHS
warmup_steps = int(total_steps * WARMUP_RATIO)

scheduler = get_cosine_schedule_with_warmup(
    optimizer,
    num_warmup_steps=warmup_steps,
    num_training_steps=total_steps
)

scaler = torch.cuda.amp.GradScaler(enabled=(DEVICE == "cuda"))


# ============================================================
# LOSS
# ============================================================

def distill_loss(student_emb, teacher_emb):
    teacher_emb = F.normalize(teacher_emb, dim=-1)

    mse = F.mse_loss(student_emb, teacher_emb)
    cosine = 1 - F.cosine_similarity(student_emb, teacher_emb, dim=-1).mean()

    loss = mse + cosine

    return loss, mse.detach(), cosine.detach()


# ============================================================
# TRAIN
# ============================================================

best_loss = float("inf")
train_losses = [] # Lưu loss để vẽ biểu đồ

for epoch in range(NUM_EPOCHS):
    student.train()

    running_loss = 0.0
    running_mse = 0.0
    running_cos = 0.0

    pbar = tqdm(train_loader, desc=f"Epoch {epoch + 1}/{NUM_EPOCHS}")

    for step, batch in enumerate(pbar):
        input_ids = batch["input_ids"].to(DEVICE, non_blocking=True)
        attention_mask = batch["attention_mask"].to(DEVICE, non_blocking=True)
        teacher_emb = batch["teacher_embedding"].to(DEVICE, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)

        with torch.cuda.amp.autocast(enabled=(DEVICE == "cuda")):
            student_emb = student(input_ids, attention_mask)
            loss, mse, cos = distill_loss(student_emb, teacher_emb)

        scaler.scale(loss).backward()

        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(student.parameters(), GRAD_CLIP)

        scaler.step(optimizer)
        scaler.update()
        scheduler.step()

        running_loss += loss.item()
        running_mse += mse.item()
        running_cos += cos.item()

        avg_loss = running_loss / (step + 1)
        avg_mse = running_mse / (step + 1)
        avg_cos = running_cos / (step + 1)

        pbar.set_postfix({
            "loss": round(avg_loss, 4),
            "mse": round(avg_mse, 4),
            "cos": round(avg_cos, 4),
            "lr": scheduler.get_last_lr()[0],
        })

    epoch_loss = running_loss / len(train_loader)
    train_losses.append(epoch_loss)

    save_path = OUTPUT_DIR / f"student_epoch_{epoch + 1}.pt"

    model_to_save = student.module if isinstance(student, nn.DataParallel) else student

    torch.save(
        {
            "epoch": epoch + 1,
            "model_state_dict": model_to_save.state_dict(),
            "tokenizer_name": PHOBERT_NAME,
            "max_length": MAX_LENGTH,
            "embedding_dim": teacher_text_embeddings.shape[1],
            "loss": epoch_loss,
            "config": {
                "phobert_name": PHOBERT_NAME,
                "batch_size": BATCH_SIZE,
                "lr": LR,
                "freeze_phobert": FREEZE_PHOBERT,
            }
        },
        save_path
    )

    print(f"[DONE] Saved epoch checkpoint: {save_path}")

    if epoch_loss < best_loss:
        best_loss = epoch_loss

        best_path = OUTPUT_DIR / "student_best.pt"

        torch.save(
            {
                "epoch": epoch + 1,
                "model_state_dict": model_to_save.state_dict(),
                "tokenizer_name": PHOBERT_NAME,
                "max_length": MAX_LENGTH,
                "embedding_dim": teacher_text_embeddings.shape[1],
                "loss": epoch_loss,
                "config": {
                    "phobert_name": PHOBERT_NAME,
                    "batch_size": BATCH_SIZE,
                    "lr": LR,
                    "freeze_phobert": FREEZE_PHOBERT,
                }
            },
            best_path
        )

        print(f"[BEST] Saved best model: {best_path}")


print("\nTraining done.")
print("Best loss:", best_loss)
print("Output dir:", OUTPUT_DIR)


# ============================================================
# PLOT LOSS
# ============================================================

plt.figure(figsize=(10, 6))
plt.plot(range(1, NUM_EPOCHS + 1), train_losses, marker='o', linestyle='-', color='b', linewidth=2, markersize=8)
plt.title('Training Loss over Epochs', fontsize=14)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.xticks(range(1, NUM_EPOCHS + 1))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

plot_path = OUTPUT_DIR / "training_loss_plot.png"
plt.savefig(plot_path)
plt.show()

print(f"[DONE] Saved loss plot to: {plot_path}")
```
